```python
import pandas as pd
import matplotlib.pyplot as plt




def people_count():
    count=0
    for u in data["User ID"]:
        count = 1 + count
        
    print("Total number of prople : ",count)
    


def purchase_count():
    pcount = 0
    dcount = 0
    for p in data["Purchased"]:
        if p is 1:
            pcount= 1+pcount
            
        elif p is 0:
            dcount= 1+dcount
    
    #pie chart
    labels= 'Purchased','Not Purchased'
    sizes = [pcount,dcount]
    colors= ['gold','blue']
    
    
    plt.pie(sizes,labels=labels,colors=colors,explode=(0.05,0),autopct='%1.1f%%',startangle=90)
    plt.axis('equal')
    plt.title("SALE Perce")
    plt.show()
    #print("Number of people who purchased :",pcount)
    #print("Number of people who didn't purchased :",dcount)
    
        



def age_dict():
    lg = max(data['Age'])
    sm = min(data['Age'])
    gap=[]
    index=0
    ulst=[]
    agedict={}
    barvalues=[]
    for x in range(sm,lg,10):
       gap.append(x) 
    
    for i in gap:
        
        if index is not len(gap)-1:
                for age in data['Age']: 
                    if age>= i and age<gap[index+1]:
                        ulst.append(data.loc[age,'User ID'])


                agedict[str(i)+'-'+str(gap[index+1])]= ulst
                barvalues.append(len(ulst))
                ulst=[]

        else:
            for age in data['Age']: 
                    if age>= i:
                        ulst.append(data.loc[age,'User ID'])
            
            
            barvalues.append(len(ulst))
            agedict[str(i)+'>']= ulst
            ulst=[]
        
        index= 1+index
        
        
    #print(agedict)
    
    # printing bar graph
    
    labels= list(agedict.keys())
    
    plt.bar(range(len(labels)),barvalues,color = ('b','g','k','r',))
    plt.xticks(range(len(labels)),labels)
    plt.title("Age Groups")
    plt.xlabel("Groups")
    plt.ylabel("No. of People")
    plt.show()
                            

 
    
def gender_ratio():
    male=0
    female=0
    for item in data['Gender']:
        if item == "Female":
             female= female+1
        else:
            male=male+1
    
    gratio= male/female*100
    print("Gender Ratio (male/female) :",gratio)
    
    #pie chart
    labels= 'Men','Women'
    sizes = [male,female]
    colors= ['yellow','pink']

    plt.pie(sizes,labels=labels,colors=colors,explode=(0.05,0),autopct='%1.1f%%',startangle=95,shadow= 'true')
    plt.axis('equal')
    plt.title("Number of Men/Women ")
    plt.show()

    

 
      

def salary():
    slst=[]
    count=0
    for sal in data['EstimatedSalary']:
        slst.append(sal)
        count=count+1
    maxsal= int(max(slst))
    minsal= int(min(slst))
    
    mean = sum(slst)/count
        
    
    print("MEAN SALARY : ",mean,"\nMAX SALARY : ",maxsal,"\nMIN SALARY : ",minsal)
    
  


def main():
    
    fpath= "./file/Social_Network_Ads.csv"
    data = pd.read_csv(fpath)
    people_count()
    purchase_count()
    age_dict()
    gender_ratio() 
    salary()
    
    
main()
```
