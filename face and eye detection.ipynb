```python
import numpy as np
import cv2 as cv

face_cascade = cv.CascadeClassifier('./file/xml/haarcascade_frontalface_default.xml')

eye_cascade = cv.CascadeClassifier('./file/xml/haarcascade_eye.xml')

cap = cv.VideoCapture(0)
```

```python
while(cap.isOpened()):
    
        ret,frame= cap.read()
        img = cv.cvtColor(frame,cv.COLOR_BGR2RGBA)

        gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

        faces= face_cascade.detectMultiScale(gray,1.3,5)

        print("Faces",faces)


        for (x,y,w,h) in faces:
            cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

            roi_gray = gray[y:y+h,x:x+w]
            roi_color= img[y:y+h,x:x+w]

            eyes = eye_cascade.detectMultiScale(roi_gray)
            print("Eyes",eyes)

            for (ex,ey,ew,eh) in eyes:
                cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)


        cv.imshow("Press q to quit",img)
        

        if cv.waitKey(1) & 0xFF == ord('q'):
                break

cv.destroyAllWindows()
```
